```python
import numpy as np
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import os
print(os.getcwd())
import os

# Load the dataset
bike_data = pd.read_csv("Bicycle_Thefts_Open_Data.csv")
```

```python
# Create a new dataset with the specified modifications
if 'STATUS' in bike_data.columns:
    # Update the STATUS column
    bike_data['STATUS'] = bike_data['STATUS'].replace({
        'STOLEN': 'NOT RECOVERED',
        'UNKNOWN': 'NOT RECOVERED',
        'RECOVERED': 'RECOVERED'
    })

    # Drop unnecessary columns
    new_bike_data = bike_data.drop(columns=['EVENT_UNIQUE_ID', 'OBJECTID', 'PRIMARY_OFFENCE'], errors='ignore')

    # Save the modified dataset to a new CSV file
    new_file_path = '/mnt/data/Bicycle_Thefts_Modified.csv'
    new_bike_data.to_csv(new_file_path, index=False)
    
    print("Updated dataset saved successfully!")
    print(f"File path: {new_file_path}")
else:
    print("The 'STATUS' column is missing from the dataset.")
```
